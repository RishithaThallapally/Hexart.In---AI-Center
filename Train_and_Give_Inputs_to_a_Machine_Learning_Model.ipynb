```python
#reading of dataset
import pandas as pd
df=pd.read_csv("heart.csv")
df
```

## data preparation

```python
#output of the dataset
y=df['output']
y
```

```python
#seperation of output from the original dataset
x=df.drop('output',axis=1)
x
```

## data splitting

```python
#splitting of complete dataset in train(80%) and test(20%)

from sklearn.model_selection import train_test_split


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_test
```

```python
x_train
```

## **model building**

# linear regression

```python
#usinng linear regression algorithm
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
```

training

```python
#assigning the predicted values
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

evaluating

```python
#original output before training
y_train
```

```python
#output after training
y_lr_train_pred
```

```python
#evaluating of results from the model
from sklearn.metrics import mean_squared_error,r2_score

lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)

lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
#results
print('LR MSE (TRAIN):',lr_train_mse)
print('LR MSE (TRAIN):',lr_train_r2)
print('LR MSE (TEST):',lr_test_mse)
print('LR MSE (TEST):',lr_test_r2)
```

```python
#results in tabular coloumn
lr_results=pd.DataFrame(["Linear regression",lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
lr_results
```

data visualization


```python
#visualizing of results
import matplotlib.pyplot as plt

plt.figure(figsize=[5,5])
plt.scatter(x=y_train,y=y_lr_train_pred,c='#7CAE00',alpha=0.3)


plt.plot()
plt.ylabel('Predicted output')
plt.xlabel('experimental output')
```
